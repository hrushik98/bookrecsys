# Creating a collaborative based RS

```python
x= ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_like_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_like_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = y[y].index
```

```python
final_ratings  = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating')
```

```python
pt.shape
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
def recommend(book_name):
    find = np.where(pt.index==book_name)[0][0];
    #distances = similarity_score[find];
    similar = sorted(list(enumerate(similarity_score[find])), key = lambda x:x[1], reverse = True)[1:6];
    for i in similar:
        print(pt.index[i[0]])
```

```python
recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```
